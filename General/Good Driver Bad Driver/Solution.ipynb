```python
import pandas as pd
```

```python
#importing data in dataset
df=pd.read_csv('drivertrainlabeled.csv')
df
```

```python
df['Driving_Class'].replace({'Good':0, 'Average':1, 'Bad':2},inplace=True)
df.head(1)
```

```python
#dividing dataframe into input and output values
dfinput= df.iloc[:,[1,2]]
dfoutput=df['Driving_Class']
```

```python
#being the case of classification and both the input variables are numeric, we can go for KNN algo for creating model
from sklearn.neighbors import KNeighborsClassifier
```

```python
#Task-1
#initailizing model
knn=KNeighborsClassifier()
#training model
knn.fit(dfinput,dfoutput)
#evaluting model
knn.score(dfinput,dfoutput)
```

```python
#getting predicted values to get confusion matrix for future evalution if need, not mendatory
df['Pre_Driving_Class']=knn.predict(dfinput)
df.head()
```

```python
actual=df['Driving_Class']
predicted=df['Pre_Driving_Class']
```

```python
from sklearn.metrics import classification_report
```

```python
print(classification_report(actual,predicted))
```
